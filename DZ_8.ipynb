```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import RidgeCV
from sklearn.svm import SVR
from sklearn.ensemble import StackingRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# видим много пустых значений, посмотрим, что внутри
df.describe()
```

```python
df.info()
```

```python
# Есть колонки с явным недостатком данных: PoolQC, MiscFeature, Alley. 
# Суммарно их почти 10% от данных, не буду выкидывать, тем более что деревья хорошо справляются с пропусками
```

```python
# Заменим пропуски
cat_col = [i for i in df.columns if df[i].dtype.name == 'object']
num_col   = [i for i in df.columns if df[i].dtype.name != 'object']

for i in cat_col:
    df[i].fillna('None', inplace=True)
    
for i in num_col:
    df[i].fillna(0, inplace=True)
    
# проверим, что не осталось пропусков    
df.info()
```

```python
# С пустотами разобрались, теперь категориальные признаки переведем в числовые
LE = LabelEncoder()
for i in cat_col:        
    df[i] = LE.fit_transform(df[i])
```

```python
# Делим данные для обучения
X = df.drop(['SalePrice'], axis=1)
y = df['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Случайный лес
rfr = RandomForestRegressor(n_estimators=100).fit(X_train, y_train)
```

```python
# Определим какие параметры важны
importances = pd.Series(rfr.feature_importances_, index=X_train.columns)
importances = importances.sort_values(ascending=False)
plt.subplots(figsize=(10, 15))
sns.barplot(x=importances, y=importances.index)
plt.xlabel('Важность')
plt.ylabel('фича')
plt.title("Важность фич")
plt.show()
```

```python
print(f'Качество random forest на тестовых данных: {rfr.score(X_test, y_test)}')
```

```python
# Стекинг
stack = StackingRegressor(estimators=[
        ('svr', SVR()),
        ('rifge', RidgeCV()),
        ('tree', DecisionTreeRegressor())] 
        ).fit(X_train, y_train)
```

```python
print(f'Качество stacking на тестовых данных: {stack.score(X_test, y_test)}')
```
